```julia
function Orthogonalize(V, ix, indices)
    # Orthogonalizes V[:, ix] against vectors V[:, indices]
    # Input:
    #           - V : Current eigenvector basis approximation
    #           - ix : Index of vector to be orthogonalized
    #           - indices: Indices of vectors against which to orthogonalize
    s = V[:, indices]'*V[:,ix]
    V[:,ix] = V[:,ix] - V[:,indices]*s
end

function Lanczos(A, H, V, k::Int, p::Int, tol, roFlag::Int, fullFirst::Bool)
    # A   - The matrix whose eigenpairs we're interested in
    # H   - The (k+1)xk Krylov-Schur matrix
    # V   - The nx(k+p+1) current Arnoldi Basis
    # k   - Current number of basis vectors
    # p   - Number of steps of Arnoldi to perform
    # tol - Required orthogonality between vectors
    # roFlag - Flag to determine what kind of reorthogonalization to do
    #            - 0 - None
    #            - 1 - Full
    #            - 2 - Partial
    #   A*V[:,1:k] = V[:,1:k+1]*H[1:k+1,1:k]
    
    n = size(A,1)  
    forceFull = false
    
    for j=1:p  
        if(j==1 && fullFirst)
            Arnoldi(A, H, V, k, 1)
            continue
        end
        
        if (k+j-1==1 || abs(H[k+j-1,k+j-2]) < tol)
            V[:,k+j-1] = rand(n,1)#copy(A[:,k+j-1])
            
            if (k+j-1 > 1)
                Orthogonalize(V, k+j-1, 1:k+j-2)
                Orthogonalize(V, k+j-1, 1:k+j-2)
            end    
            
            V[:,k+j-1] /= norm(sub(V, 1:n, k+j-1))
        end
        
        w = A*V[:,k+j-1]
        if (k+j-1 > 1)
            h = V[:,k+j-2:k+j-1]'*w
            H[k+j-2:k+j-1, k+j-1] = h
            V[:,k+j] = w - V[:,k+j-2:k+j-1]*h
        else
            h = V[:,k+j-1:k+j-1]'*w
            # Julia doesn't convert 1x1 matrices into scalars
            H[k+j-1:k+j-1, k+j-1:k+j-1] = h
            V[:,k+j] = w - V[:,k+j-1:k+j-1]*h
        end
        beta = norm(V[:,k+j])
        if (beta > tol)
            V[:,k+j] = V[:,k+j]/beta
            H[k+j, k+j-1] = beta
        else
            H[k+j, k+j-1] = 0
        end
        if (roFlag == 1 || (forceFull == true))
            # Perform full-orthogonalization
            Orthogonalize(V, k+j, 1:(k+j-1))       
            Orthogonalize(V, k+j, 1:(k+j-1))
            
            if (roFlag == 2)
                forceFull = false
            end
        elseif (roFlag == 2)
            # Perform selective orthogonalization
        end
    end
end

function UpdateOmega(omega, omegaOld, a, b, j)
    eps = 1e-16;
    for i=1:j-1
        if (i-1 == 0)
            wkm1 = 0
            bim1 = 0
        else
            wkm1 = omega[i-1]
            bim1 = b[i-1]
        end
        
        omegaOld[i] = (b[i]*omega[i+1] + (a[i] - a[j])*omega[i] + bim1*wkm1 - b[j-1]*omegaOld[i])/b[j] + 1e-16
    end

    omegaOld[j] = eps;
    omegaOld[j+1] = 1;
end

function LanczosTriDiag(A, a, b, V, k::Int, p::Int, tol, roFlag::Int, fullFirst::Bool)
    # A   - The matrix whose eigenpairs we're interested in
    # a   - Diagonal of Tridiagonal Rayleigh Quotient
    # b   - Subdiagonal of Tridiagonal Rayleigh Quotient
    # V   - The nx(k+p+1) current Arnoldi Basis
    # k   - Current number of basis vectors
    # p   - Number of steps of Arnoldi to perform
    # tol - Required orthogonality between vectors
    # roFlag - Flag to determine what kind of reorthogonalization to do
    #            - 0 - None
    #            - 1 - Full
    #            - 2 - Partial
    #   A*V[:,1:k] = V[:,1:k+1]*H[1:k+1,1:k]

    n = size(A,1)  
    forceFull = false
    
    if (roFlag == 2)
        omegaIndices = 0
        omega = zeros(k+p)
        omegaOld = zeros(k+p)
        omega[1] = 1;
    end
    for j=1:p        
        if (k+j-1==1 || abs(b[k+j-2]) < tol)
            V[:,k+j-1] = rand(n,1)#copy(A[:,k+j-1])
            
            if (k+j-1 > 1)
                Orthogonalize(V, k+j-1, 1:k+j-2)
                Orthogonalize(V, k+j-1, 1:k+j-2)
            end
            
            V[:,k+j-1] /= norm(sub(V, 1:n, k+j-1))
        end
        
        w = A*V[:,k+j-1]
        if (k+j-1 > 1)
            h = V[:,k+j-2:k+j-1]'*w
            a[k+j-1] = h[2]
            b[k+j-1] = h[1]
            V[:,k+j] = w - V[:,k+j-2:k+j-1]*h
        else
            h = V[:,k+j-1:k+j-1]'*w
            a[k+j-1] = h[1]
            V[:,k+j] = w - V[:,k+j-1:k+j-1]*h
        end
        beta = norm(V[:,k+j])
     
        if (beta > tol)
            V[:,k+j] = V[:,k+j]/beta
            b[k+j-1] = beta
        else
            b[k+j-1] = 0;
            continue
        end
        
        if (roFlag == 1)
            # Perform full-orthogonalization
            Orthogonalize(V, k+j, 1:(k+j-1))       
            Orthogonalize(V, k+j, 1:(k+j-1))
            
        elseif (roFlag == 2 && forceFull)
            # Forced orthogonalization after partial is performed
            Orthogonalize(V, k+j, indices)       
            Orthogonalize(V, k+j, indices)
            beta = norm(V[:,k+j])
            V[:,k+j] /= beta
            b[k+j-1] *= beta
            
            UpdateOmega(omega, omegaOld, a, b, k+j-1)
            omega, omegaOld = omegaOld, omega
            
            omega[indices] = 1e-16
            omega[k+j-1] = 1e-16
            omega[k+j] = 1
            
            forceFull = false
            
        elseif (roFlag == 2)
            # Perform partial reorthogonalization
            UpdateOmega(omega, omegaOld, a, b, k+j-1)
            omega, omegaOld = omegaOld, omega # No good way to swap pointers inside of UpdateOmega
            
            omegaMax, z = findmax(omega[1:k+j-1])
            
            if (omegaMax > 1e-10/sqrt(k+j-1))  

                ix = find(x -> (x > 1e-12/sqrt(k+j-1)), omega[1:k+j-1])
                r,z = findmin(ix)
                s,z = findmax(ix)
                
                indices = r:s#1:k+j-1#
                
                Orthogonalize(V, k+j, indices)                
                Orthogonalize(V, k+j, indices)
                
                beta = norm(V[:,k+j])
                V[:,k+j] /= beta
                b[k+j-1] *= beta

                forceFull = true

                omega[indices] = 1e-16
            end
        end
    end
end

function Arnoldi(A, H, V, k::Int, p::Int)
    # A   - The matrix whose eigenpairs we're interested in
    # H   - The (k+1)xk Krylov-Schur matrix
    # V   - The nx(k+p+1) current Arnoldi Basis
    # k   - Current number of basis vectors
    # p   - Number of steps of Arnoldi to perform
    #   A*V[:,1:k] = V[:,1:k+1]*H[1:k+1,1:k]
    
    n = size(A,1)
    eps = 1e-12
    
    for j=1:p
        if (k+j-1==1)
            V[:,k+j-1] = copy(A[:,k+j-1])
            V[:,k+j-1] /= norm(sub(V, 1:n, k+j-1))
        end
        
        w = A*V[:,k+j-1]
        h = V[:,1:k+j-1]'*w
        H[1:k+j-1, k+j-1] = h
        V[:,k+j] = w - V[:,1:k+j-1]*h
        beta = norm(V[:,k+j])
        if (beta > eps)
            V[:,k+j] = V[:,k+j]/beta
            H[k+j, k+j-1] = beta
        else
            H[k+j, k+j-1] = 0
        end
   
        #Re-orthogonolization
        s = V[:,1:k+j-1]'*V[:,k+j]
        while (norm(s) > eps*norm(V[:,k+j]) && norm(s) > eps)
            V[:,k+j] -= V[:,1:k+j-1]*s
            s = V[:,1:k+j-1]'*V[:,k+j]
        end
    end
end
    
function SchurForm(V, H, k)
    # Computes the Schur form of H = X*S*X', and updates V=V*X
    # Input: 
    #           - V : Current eigenvector basis approximation
    #           - H : Current Rayleigh value approximation
    #           - k : Number of eigenvalues we want to compute
    # Use symmetric part due to rounding errors
    S, X = eig(H[1:k,1:k])#0.5*(H[1:k,1:k] + H[1:k,1:k]'))
    V[:,1:k] = V[:,1:k]*X
    H[1:k,1:k] = diagm(S)
    H[k+1,1:k] = H[k+1,k]*X[k,:]
    
    Orthogonalize(V, k+1, 1:(k-1))       
    Orthogonalize(V, k+1, 1:(k-1))
end

function ImplicitShift(V, H, k, p, filterFunc=None)
    
    evals = eigvals(H[1:k+p,1:k+p])
    indices = PurgeShifts(evals, p, filterFunc)
    
    Q = eye(k+p);
    for i in indices
        Qj,Rj = qr(H[1:k+p,1:k+p] - evals[i]*eye(k+p))
        H[1:k+p,1:k+p] = Qj'*H[1:k+p,1:k+p]*Qj
        Q = Q*Qj
    end    
    
    v = V[:,1:k+p]*Q[:,k+1]
    V[:,1:k] = V[:,1:k+p]*Q[:,1:k]
    V[:,k+1] = v*H[k+1,k] + V[:,k+p+1]*H[k+p+1,k+p]*Q[k+p,k]
    beta = norm(V[:,k+1])
    V[:,k+1] /= beta
    H[k+1,1:k] = 0 # FIX THIS
    H[:, k+1:] = 0 # FIX THIS
    H[k+2:,:] = 0 # FIX THIS
    H[k+1,k] = beta
    
    Orthogonalize(V, k+1, 1:k)
    Orthogonalize(V, k+1, 1:k)
end

function Givens(a, b)
    # Do simple way first
    r = sqrt(a*a + b*b)
    c = a/r
    s = -b/r
    
    return c,s,r
end

function QrIterationTriDiag(a, b, shift, Q, kp)
    # Matrix will look like [b, a, bCopy, c] with b as subdiagonal, a diagonal, etc.
    a[:] -= shift*ones(kp)
    bCopy = copy(b)
    
    givensList = zeros(kp-1,2)
    
    # QR of T - shift*I
    for i = 1:kp-1
        # Get Givens rotation
        c,s,r = Givens(a[i], b[i])
        givensList[i,1] = c
        givensList[i,2] = s
        
        #Apply Givens rotation from left
        a[i] = r
        b[i] = 0

        tmp1 = c*bCopy[i] - s*a[i+1]
        tmp2 = s*bCopy[i] + c*a[i+1]
        bCopy[i] = tmp1
        a[i+1] = tmp2
        if (i < kp - 1)
            bCopy[i+1] = c*bCopy[i+1]
        end

    end
    
    # RQ product and update of Q
    for i = 1:kp-1
        c = givensList[i,1]
        s = givensList[i,2]
        
        # Update T
        tmp = a[i]
        a[i] = c*a[i] - s*bCopy[i]
        bCopy[i] = s*tmp + c*bCopy[i]
        
        b[i] = -s*a[i+1]
        a[i+1] = c*a[i+1]
        
        # Update Q
        tmp = copy(Q[:,i])
        Q[:,i] = c*Q[:,i] - s*Q[:,i+1]
        Q[:,i+1] = s*tmp + c*Q[:,i+1]
        
    end
    
    a[:] += shift*ones(kp)
end

function ImplicitShiftTriDiag(V, a, b, k, p, filterFunc=None)
    
    T = SymTridiagonal(a, b[1:k+p-1])
    evals = eigvals(T)
    
    indices = PurgeShifts(evals, p, filterFunc)
    
    Q = eye(k+p);
    for i in indices
        QrIterationTriDiag(a, b, evals[i], Q, k+p)
    end    
    
    v = V[:,1:k+p]*Q[:,k+1]
    V[:,1:k] = V[:,1:k+p]*Q[:,1:k]
    V[:,k+1] = v*b[k] + V[:,k+p+1]*b[k+p]*Q[k+p,k]
    beta = norm(V[:,k+1])
    V[:,k+1] /= beta
    b[k] = beta
    
    Orthogonalize(V, k+1, 1:k)
    Orthogonalize(V, k+1, 1:k)
    
end

function PurgeShifts(evals, p, filterFunc=None)
    # Returns indices of eigenvalues to be purged
    # Input:
    #           - evals : Current eigenvalues for 
    #           - p : Number of eigenvalues we want to purge
    #           - filterFunc : Function which takes eigenvalues and returns indices of which eigenvalues we want to purge.
    #                          If None, then purge smallest eigenvalues.
    # Output:
    #           Indices in evals of Ritz pairs we want to purge.
    if (filterFunc == None)
        return sortperm(evals)[1:p]
    else
        return filterFunc(evals, p)
    end
end

function Shifts(evals, k, filterFunc=None)
    # Returns indices of eigenvalues to be kept
    # Input:
    #           - evals : Current eigenvalues for 
    #           - k : Number of eigenvalues we want to keep
    #           - filterFunc : Function which takes eigenvalues and returns indices of which eigenvalues we want to keep.
    #                          If None, then keep largest eigenvalues.
    # Output:
    #           Indices in evals of Ritz pairs we want to keep.
    n = size(evals,1)
    if (filterFunc == None)
        perm = sortperm(evals)[n-k+1:n]
    else
        perm = filterFunc(evals, k)
    end
    
    return perm
end

function ConvergedImplicitArnoldi(H, tol, iters, maxIter)
    return (iters > maxiter) || (norm(H[k+1,k]) < tol)
end

function ConvergedImplicitArnoldiTriDiag(b, tol, iters, maxIter)
    return (iters > maxiter) || (abs(b[k]) < tol)
end

function Converged(H, k, tol, iters, maxIter)
    # Returns True if we are ready to exit the eigenvalue solver
    # Input:    
    #           - H : The (expanded) Rayleigh Quotient
    #           - k : The number of eigenvalues we want
    #           - tol : required tolerance for error vector
    #           - iters : Current number of iterations
    #           - maxIter : Max number of iterations for eigenvalue solver
    # Output:
    #           True if we've reached the max number of iterations, or
    # the 2-norm of the error vector is below tol.
    # May need to fix convergence Criteria
    return (iters > maxiter) || (norm(H[k+1,1:k]) < tol)
end

function KrylovSpectralSolver(A, k, p, tol, maxiter, roFlag, filterFunc=None)
    # Computes k eigenvalues using the implicitly restarted Lanczos algorithm
    # while storing a basis of size at most k+p.
    # Runs until all Ritz pairs (x, lambda) satisfy ||Ax - lambda*x|| < tol
    # or we reach maxiter iterations.
    # Input:
    #           - A : Matrix whose eigenvalues we want
    #           - k : Number of eigenvalues desired
    #           - p : k+p defines number of Lanczos vectors to store
    #           - tol : Tolerance for Ritz pair convergence
    #           - maxiter: Maximum number of iterations
    #           - filterFunc : Function which takes (evals, k) where
    #                  evals is a list of k+p eigenvalues, and k is the
    #                  number of eigenvalues we want to keep. filterFunc
    #                  returns the indices of the eigenvalues that we want
    #                  to keep based on some selection criteria.
    # Output:
    #           - V : The Eigenvector basis approximation
    #           - H : A (k+p+1 x k+p) matrix with the eigenvalues on
    #                 the diagonal of H[1:k,1:k]
    #           - iters: The number of iterations taken
    
    # Initialize necessary things
    n = size(A,1)
    m = k+p
    V = zeros(n,m+1)
    H = zeros(m+1,m)
    iters = 1
        
#    Arnoldi(A, H, V, 1, k)
    Lanczos(A, H, V, 1, k, 1e-8, roFlag, false)
    SchurForm(V, H, k)
    
    while (!Converged(H, k, tol, iters, maxiter))
#        Arnoldi(A, H, V, k+1, p)
        
        Lanczos(A, H, V, k+1, p, 1e-8, roFlag, true)
       
        SchurForm(V, H, k+p)  
        
        evals = diag(H[1:k+p,1:k+p])
        perm = Shifts(evals, k, filterFunc)
        
        V[:,1:k] = V[:,perm]
        V[:,k+1] = V[:,k+p+1]
        H[1:k,1:k] = diagm(evals[perm])
        H[k+1,1:k] = H[k+p+1,perm]
        
        iters+=1
    end
    
    return (V,H,iters)
end

function ImplicitlyRestartedLanczos(A, k, p, tol, maxiter, roFlag, filterFunc=None)
    # Computes k eigenvalues using the implicitly restarted Lanczos algorithm
    # while storing a basis of size at most k+p.
    # Runs until all Ritz pairs (x, lambda) satisfy ||Ax - lambda*x|| < tol
    # or we reach maxiter iterations.
    # Input:
    #           - A : Matrix whose eigenvalues we want
    #           - k : Number of eigenvalues desired
    #           - p : k+p defines number of Lanczos vectors to store
    #           - tol : Tolerance for Ritz pair convergence
    #           - maxiter: Maximum number of iterations
    #           - filterFunc : Function which takes (evals, k) where
    #                  evals is a list of k+p eigenvalues, and k is the
    #                  number of eigenvalues we want to keep. filterFunc
    #                  returns the indices of the eigenvalues that we want
    #                  to keep based on some selection criteria.
    # Output:
    #           - V : The Eigenvector basis approximation
    #           - H : A (k+p+1 x k+p) matrix with the eigenvalues on
    #                 the diagonal of H[1:k,1:k]
    #           - iters: The number of iterations taken
    
    # Initialize necessary things
    n = size(A,1)
    m = k+p
    V = zeros(n,m+1)
#    H = zeros(m+1,m)
    a = zeros(m)
    b = zeros(m)
    iters = 1
        
#    Lanczos(A, H, V, 1, k, 1e-8, roFlag, false)
    LanczosTriDiag(A, a, b, V, 1, k, 1e-8, roFlag, false)
    
    while (!ConvergedImplicitArnoldiTriDiag(b, tol, iters, maxiter))#Converged(H, k, tol, iters, maxiter))
#        Lanczos(A, H, V, k+1, p, 1e-8, roFlag, false)
        LanczosTriDiag(A, a, b, V, k+1, p, 1e-8, roFlag, false)
#        ImplicitShift(V, H, k, p, filterFunc)
        ImplicitShiftTriDiag(V, a, b, k, p, filterFunc)
        iters+=1
    end
    
    S,X = eig(SymTridiagonal(a[1:k],b[1:k-1]))
    V[:,1:k] = V[:,1:k]*X 
    
    return (V,S,iters)
end
```

```julia
function Laplacian(n)
    L = 4*speye(n*n,n*n)
    for i=1:n*n
        if (i > 1 && (i-1)%n != 0)
            L[i-1,i] = -1;
        end
        if (i < n*n && i%n != 0)
            L[i+1,i] = -1;
        end

        if (i <= n*n - n)
            L[i+n,i] = -1
        end
        if (i <= n*n - n)
            L[i,n+i] = -1
        end
    end
    return L
end
```
